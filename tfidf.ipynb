```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vect = TfidfVectorizer()

tfidf_matrix = tfidf_vect.fit_transform(df.clean_text_lemma)
```

```python
import pandas as pd
from scipy.sparse import csr_matrix

# Assuming 'tfidf_matrix_custom' is your TF-IDF matrix
# Convert the TF-IDF matrix to a sparse matrix
tfidf_sparse_matrix = csr_matrix(tfidf_matrix)

# Concatenate the original DataFrame with the TF-IDF sparse matrix along the columns
merged_dataframe_sparse = pd.concat([df, pd.DataFrame.sparse.from_spmatrix(tfidf_sparse_matrix, columns=tfidf_vect.get_feature_names_out())], axis=1)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def filter_similar_records_custom(query, dataframe, vectorizer, matrix, threshold=None, top_n=None):
    # Transform the query using the provided TF-IDF vectorizer
    query_tfidf = vectorizer.transform([query])

    # Calculate cosine similarity between the query and all documents
    cosine_similarities = cosine_similarity(query_tfidf, matrix).flatten()

    # Add cosine similarity scores as a new column in the DataFrame
    dataframe['cosine_similarity_custom'] = cosine_similarities

    # Sort the DataFrame based on cosine similarity scores
    sorted_dataframe = dataframe.sort_values(by='cosine_similarity_custom', ascending=False).reset_index(drop=True)

    # Filter based on threshold similarity score or select top N similar records
    if threshold is not None:
        filtered_dataframe = sorted_dataframe[sorted_dataframe['cosine_similarity_custom'] >= threshold]
    elif top_n is not None:
        filtered_dataframe = sorted_dataframe.head(top_n)
    else:
        filtered_dataframe = sorted_dataframe

    return filtered_dataframe
```

```python
# Example usage:
search_query = input('Enter your search query: ')
# Lowering the threshold initially to get some results
similarity_threshold = 0.1
top_records_to_retrieve = int(input('Enter the number of top similar records to retrieve: '))

filtered_results = filter_similar_records_custom(search_query, df, tfidf_vect, tfidf_matrix, threshold=similarity_threshold, top_n=top_records_to_retrieve)
print(f"Filtered records for the query '{search_query}':")
filtered_results[['name', 'num', 'cosine_similarity_custom']]
```
